```python
# trying to define my own function, this function will take a dictionary (freq_distribution) as argument 
# and return two lists or arrays, a list of rank and their corresponding frequency

def rank_freq(x):
    
    def ascd(x):
        ascending = sorted(x.items(),key = lambda x: x[1],reverse = True) # convert dict to tuple & sort the values
                                                                          # in ascending order
        return ascending
    
    ascen = ascd(x)
    
    def proc_freq(ascen): # takes the tuple from above,extracts the freq and
                          # groups the samples with equal freq
        arrange_freq = []
        frequency = [value for key,value in ascen]
        for b in sorted(list(set(frequency)),reverse=True):
            corresp = []
            for c in frequency:
                if c == b:
                    corresp.append(c)
            arrange_freq.append(tuple(corresp))
        return (arrange_freq,frequency)
    
    arr,fre = proc_freq(ascen)
            
    def rank_it(arr,fre): # takes the list of tuples and produces the list of ranks
        rank_list = []
        n = 1
        for l in arr:
            if len(l)== 1:
                rank_list.append(n)
                n+=1
            else:
                rank_list = rank_list + [float(sum(range(n,n+len(l))))/len(l)]*len(l)
                n+=len(l)
        return (rank_list,fre)
    return (rank_it(arr,fre))

v = {'A':1,'B':2,'C':3,'D':4,'E':3,'F':2,'G':3,'H':4,'I':1}
rank_freq(fdist)
            
            
```

```python
def desc_rank(x,y):
    part_desc = list()
    temp_desc = list()
    for i in x:
        if i == min(x):
            part_desc.append(i)
    temp_desc = [max(y)]*len(part_desc)
    x = filter(lambda a: a != min(x), x)
    y = filter(lambda a: a != max(y), x)
    return (temp_desc,x,y)

```

```python
# we implement our own version of the tfidf func.
# it is broken down into five different functions


# The first function counts the number of time, a word appears across a document. We used the principle of set
# theory to achieve this, e.g an intersection of all the pages returns a list of words, and these words the words
# appeared [len(no of pages)] times. we remove these words from the original set by finding the set difference
# between them.

# for example is we have the following set [[a,b,c,c],[t,w,h,a],[o,o,r,s]], it can be seen that there are three 
# three (in our case). A word that appear across all pages will be the intersection of the list elements. 

# we proceed by finding the word that appears in just 2 pages(still using the example). to achieve this, we used
# the itertools to get the list of possible number of combinations of two pages using the list index. we do this
# continously until we get to the words that appears just once. I CALLED THE FUNCTION MUST_WORK, BECAUSE IT TOOK A
# it reminds me of the time it took. it will return two values: a dictionary of the number of occurences as key and
# corresponding words as values, then the total number of pages.

def mustwork(x):
    printout = " at least two documents...aahh i meant a dict of length = 2"
    coun = 1
    x = [set(e) for e in x.values()]
    doc_len = len(x)

    
    if doc_len >= 2:
        idf_dict = {k: set([]) for k in range(1, doc_len + 1)}
        perm = [bb for ii in range(doc_len) for bb in itertools.combinations(range(doc_len),ii)][1:]
        idf_dict[doc_len] |= set(x[0]).intersection(*x)
        x = [e-idf_dict[doc_len] for e in x]
        
        
        while True:
            if coun!= doc_len - 1:
                coco = doc_len - coun
                tem = [jj for jj in perm if len(jj)==coco]
                for i in tem:
                    lisa = [x[w] for w in i]
                    idf_dict[coco] |= lisa[0].intersection(*lisa)
                x = [e-idf_dict[coco] for e in x]
                coun+=1
            else:
                break
        
        idf_dict[1] |= set([ev for every in x for ev in list(every)])
        
        return (doc_len,idf_dict)
    else:
        return printout
```

```python
# we now calculate the idf with this function

def idf(x):
    doc_len,y = mustwork(x)
    idf_classified = {k:[] for k in x.keys()}
    
    for keys,values in y.items():
        for kw,vw in x.items():
            kissme = list(set(vw).intersection(values))
            count = math.log10(doc_len/keys)
            idf_classified[kw].extend([(i,count) for i in kissme])
            
    return idf_classified
```

```python
# we get the tf with this function

def tf_mustwork(x):
    word_tf = {k:[] for k in x.keys()}
    comb_dup = {k:[] for k in x.keys()}
        

    for keys,values in x.items():
        comb_dup[keys] = Counter(values)
        
    for keys,values in comb_dup.items():
        maxi = max([num for wd,num in values.items()])
        
        for w,n in values.items():
            n = 0.5+(0.5*(n/maxi))
            word_tf[keys].append((w,n))
            
    return word_tf         
```

```python
# we multiply tf and idf to get tfidf, we simply sorted the both list of tuples(first element of both tuples) to 
# match

def tfidf(x):
    y = idf(x)
    term_frk = tf_mustwork(x)
    tfidf_dict = {q:[] for q in y.keys()}
    y = {k:list(y[k]) for k in y.keys()}
    
    for keys in y.keys():
        a = sorted(y[keys], key=lambda tup: tup[0])
        b = sorted(term_frk[keys], key=lambda tup: tup[0])
        tfidf_dict[keys]=[(c[0],c[1]*d[1]) for c,d in zip(a,b) if c[0]==d[0]]
    return tfidf_dict
```

```python
# we convert to a vector format

def vectorizing(x):
    gud = tfidf(x)
    bor = list(set([x for xx in x.values() for x in xx]))
    vect = []
    
    
    for bb in gud.keys():
        freak = list(set(bor)-set(dict(gud[bb]).keys()))
        freak = gud[bb] + [(ff,0) for ff in freak]
        freak = sorted(freak, key=lambda tup: tup[0])
        vect.append([k for i,k in freak])
    
    return vect
```
